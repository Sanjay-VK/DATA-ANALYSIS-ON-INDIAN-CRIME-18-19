```python
import pandas as pd
# Load the dataset
data = pd.read_csv("C:/Users/Hp/Desktop/INDIA CRIME/updated data set.csv")
```

```python
data.head(2)
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
import matplotlib.pyplot as plt

# Bar chart for total crimes by city
plt.figure(figsize=(12, 6))
plt.bar(data['City'], data['Total Crimes'], color='lightgreen')
plt.xlabel('City')
plt.ylabel('Total Crimes')
plt.title('Total Crimes by City')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Aggregate total crimes by region
region_crimes = data.groupby('Region')['Total Crimes'].sum()

# Pie chart for total crimes by region
plt.figure(figsize=(8, 8))
plt.pie(region_crimes, labels=region_crimes.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Total Crimes by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Aggregate total crimes by state
state_crimes = data.groupby('State/UT')['Total Crimes'].sum().reset_index()
import matplotlib.pyplot as plt

# Create a column chart for total crimes by state
plt.figure(figsize=(12, 6))
plt.bar(state_crimes['State/UT'], state_crimes['Total Crimes'], color='Green')
plt.xlabel('State/UT')
plt.ylabel('Total Crimes')
plt.title('Total Crimes by State/UT')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Find the state with the highest total crimes
highest_crime_state = state_crimes.loc[state_crimes['Total Crimes'].idxmax()]
print(f"The state with the highest total crime rate is {highest_crime_state['State/UT']} with {highest_crime_state['Total Crimes']} total crimes.")
```
